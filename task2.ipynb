```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("Iris.csv")

df.head()
```

```python
print("Shape:", df.shape)

df.info()
```

```python
df.isnull().sum()
```

```python
species_count = df["Species"].value_counts()

plt.figure(figsize=(6,4))
species_count.plot(kind="bar")

plt.title("Number of Flowers in Each Species")
plt.xlabel("Species")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.hist(df["SepalLengthCm"], bins=10)

plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,6))

for species in df["Species"].unique():

    subset = df[df["Species"] == species]

    plt.scatter(
        subset["SepalLengthCm"],
        subset["PetalLengthCm"],
        label=species
    )

plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend()

plt.show()
```

```python
avg_features = df.groupby("Species")[
    [
        "SepalLengthCm",
        "SepalWidthCm",
        "PetalLengthCm",
        "PetalWidthCm"
    ]
].mean()

avg_features
```

```python
avg_features.plot(
    kind="bar",
    figsize=(10,6)
)

plt.title("Average Feature Values by Species")
plt.xlabel("Species")
plt.ylabel("Average Value")

plt.show()
```

Conclusion:

1. Each species contains 50 samples.
2. Sepal Length values are distributed between approximately 4 cm and 8 cm.
3. Setosa is clearly separated from other species in the scatter plot.
4. Versicolor and Virginica show some overlap.
5. Petal measurements provide better separation among species.
